```python
import openpyxl
import pandas as pnd

def read_excel_table(filename, nrows):
    book = openpyxl.load_workbook(filename = filename, read_only = True, data_only = True)
    first_sheet = book.worksheets[0]
    rows_generator = first_sheet.values
    
    header_row = next(rows_generator)
    data_rows = [row for(_, row) in zip(range(nrows - 1), rows_generator)]
    return pnd.DataFrame(data_rows, columns = header_row)

churn_dataset = read_excel_table('turknetchurnekimanonim.xlsx', nrows = 192295)
churn_dataset
```

```python
import statemodels.api as smod

train_columns = list(set(list(churn_dataset)) - set(['DURUM']))
looked_column = list(set(['DURUM']))

X = churn_dataset[train_columns].T.drop_duplicates().T
y = churn_dataset[looked_column]

cols = list(X.columns)
pmax = 1 
while (len(cols>0)):
    p = []
    X_1 = X[cols]
    X_1 = smod.add_constant(X_1)
    model = smod.OLS(y, X_1).fit()
    p = pnd.Series(model.pvalues.values[1:].index = cols)
    pmax = max(p)
    feature_with_p_max = p.idxmax()
    if(pmax > 0.05):
        cols.remove(feature_with_p_max)
    else:
        break
        
selected_features = cols
print(selected_features)
len(selected_features)
    
```
